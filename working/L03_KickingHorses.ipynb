# Kicking Horses
We want to show which is the distribution of the men kicked to death in the Prussian army. Since we ar talking about counting rare events, we suppose that the underlying distribution is a Poisson distribution, with the mean given by the mean of the sample.

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
xbins = [-0.5, 0.5, 1.5, 2.5, 3.5, 4.5]
corps = [109, 65, 22, 3, 1]

data = np.zeros(0)
for i in range(0, 5):
    data = np.append(data, np.full(corps[i], i))

plt.hist(data, bins=xbins, density=True, label = 'data')
distPoiss = sp.stats.poisson(np.mean(data))
x = np.arange(-1, 5, 1)
plt.plot(x, distPoiss.pmf(x), color='orange', marker = 'o', label='Poisson distr.')
plt.legend()
plt.xlim(-0.5, 5)
```

As the plot show, we can say that (at least qualitatively), the data follow the Poisson distribution.
