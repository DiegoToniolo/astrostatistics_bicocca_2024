# Probability distribution of the log of a variable uniformly distributed
This is an example of changing of the probability distribution when you change variable. First, we draw a sample $x$ from the uniform distribution between 0.1 and 10:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100000
x = np.random.uniform(0.1, 10.0, N)
_ = plt.hist(x, bins=50, density = True)
```

Then we compute the $\log_{10}$ of the sample. From the rule for the change of variable, we know that the distribution of $y$ is:
$$p(y) = \biggl|\frac{dx}{dy}\biggr|p(x(y)) = \frac{\log{10}}{10-0.1} 10^y, \qquad y \in [\log_{10}{0.1},\log_{10}{10}] = [-1, 1]$$

```python
y = np.empty(N)
for i in range(0, N):
    y[i] = np.log10(x[i])

x_ax = np.arange(-1, 1, 0.0002)
y_ax = np.empty(10000)
for i in range(0, 10000):
    y_ax[i] = np.log(10)/9.9*10.0**(x_ax[i])

plt.hist(y, 50, density=True, label="Data")
plt.plot(x_ax, y_ax, 'r-', linewidth=2.0, label="Expected distribution")
plt.legend()
```

```python
print("Log of the mean of x: {:.3f}".format(np.log10(np.mean(x))))
print("Mean of the log of x: {:.3f}".format(np.mean(y)))
```

```python
print("Log of the median of x: {:.3f}".format(np.log10(np.median(x))))
print("Median of the log of x: {:.3f}".format(np.median(y)))
```
